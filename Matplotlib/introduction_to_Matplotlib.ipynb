# Introduction Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
plt.plot(x, y)
plt.show()
```

```python
# 1st method
fig = plt.figure() # ceates a figure
ax = fig.add_subplot() # adds some axes
plt.show()
```

```python
# 2nd method
fig = plt.figure() # creates a figure
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y) # add some data
plt.show()
```

```python
# 3rd method (recommendet)
fig, ax = plt.subplots()
ax.plot(x, [50, 100, 200, 350]) # ass some data
plt.show()
```

### MAtpotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 40, 50]

# 2. Setup plot
fig, ax = plt.subplots(figsize = (10, 7)) #width and height

# 3. plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title = "Simple Plot", xlabel = "X axis", ylabel = "Y axis" )

# 5. Save & show (you save the whole figure)
fig.savefig('sample-plot.png')
plt.show()
```

# Scatter Plot And Bar plot

We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
# Creat some data
x = np.linspace(0, 10, 100)
x
```

```python
# Plot the data
fig, ax = plt.subplots() 
ax.plot(x, x**2);
```

```python
# Use same data to make a scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionar
nut_butter_prices = {"Almound butter" : 10, "Beanut butter" : 8, "Cashew butter" : 12}

fig, ax = plt.subplots()

ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title = "Butters prices", ylabel = "Price $");
```

# Histograms And Subplots

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Two options for subplots

```python
# Sublot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, 
                                             ncols = 2,
                                             figsize = (10, 9))

# Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize = (10, 9))

ax[0, 0].plot(x, x**2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000));
```
